# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

## 1.Gender Distribution: Analyzing the Count of Genders

```python
ax = sns.countplot(x = 'Gender',data = salesdata)

for bars in ax.containers:
    ax.bar_label(bars)
```

- Observation: In the given data or sample population, the gender distribution indicates a notable difference in the number of  femalesand males. Specifically, the count of females surpasses that of females.

----------------------

## 2.Gender-Based Spending Patterns: Understanding Financial Trends Across Genders

```python
sales_gen = salesdata.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

- Observation:It is observed that female participants tended to spend more compared to their male counterparts.

-------

## 3.Demographic Distribution: Count of Age Segments by Gender

```python
ax = sns.countplot(data = salesdata, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

- Observation:The demographic distribution analysis reveals an interesting trend wherein the age segment between 26 and 35 exhibits a significantly higher count compared to other age segments across genders.

-----------

## 4.Gender-Based Spending Patterns Across Age Segments: Analyzing Consumer Behavior

```python

sales_age = salesdata.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)

```

- Observation: The demographic distribution analysis reveals that the age segment 26-35 exhibits the highest spending compared to other age segments, regardless of gender.

--------

## 5.Top 10 States with the Highest Orders: A Closer Look at Regional Consumer Trends

```python
sales_state = salesdata.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

- Observation:Urban centers, such as Uttar Pradesh, Maharashtra & Karnataka stand out as the primary contributors to the highest order volumes

-------

## 6.Top 10 States with the Highest Spending : A Closer Look at Regional Consumer SpendingTrends

```python
sales_state = salesdata.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

- Observation: Uttar Pradesh, Maharashtra & Karnataka cities boast higher population densities and more significant economic activities, leading to increased consumer spending.

----

## 7. An Analysis of Marital Status:Comparison of Married and Unmarried Individuals.

```python
ax = sns.countplot(data = salesdata, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

- Observation:The count of unmarried users is significantly higher than that of married users.

--------

## 8. Comparing Spending Patterns: Married vs. Unmarried Individuals by Gender

```python
sales_state = salesdata.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

- Observation:It is observed that Married & Unmarried female participants tended to spend more compared to their Married & Unmarried male counterparts.

--------

## 9.Occupation Diversity: Exploring a Wide Range of User Professions.

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = salesdata, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

- Observation: The data analysis reveals that among the listed occupations, namely IT, Healthcare, Aviation, and Banking, the highest number of individuals have diverse occupations.

----------------

## 10.Analyzing Occupation-Based Spending Patterns: A Comprehensive Overview

```python
sales_state = salesdata.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

- Observation: Based on the data analysis, it is evident that individuals with occupations in IT, Healthcare, Aviation, and Banking exhibit the highest spending behavior.


----------

## 11.Bestselling Product Category: Unveiling the Most Popular Orders

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = salesdata, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

- Observation: After analyzing the orders data and customer preferences, it is evident that the bestselling product category is food, followed by Clothing & Apparel, Electronics & Gadgets and Foot ware. The popularity of these categories is evident from the frequency of orders placed by customers.

-----

## 12.Total Sales Amount by Product Category

```python
sales_state = salesdata.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

- Observation: Upon analyzing the sales data, it is evident that the four categories, namely food, Clothing & Apparel, Electronics & Gadgets, and Footwear, are highly popular among customers. The total sales amount for each category indicates their significant impact on consumer spending and preferences.
